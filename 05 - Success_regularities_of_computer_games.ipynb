```python
Games.info()
```



Таблица содержит 16715 строк.


Столбец Year_of_Release (год выпуска) - дискретная переменная, приведем ее к типу int.


```python
# приведение дискретных переменных к целочисленному типу
Games['year_of_release'] = Games['year_of_release'].astype('Int64')
Games['year_of_release'].dtype #Проверка
```

Пустые строки critic_score и user_score, удалять не будем. Так как много пропущенных, они изменят общую статистическую картину и мы сделаем не правильные выводы. Оценки могут быть пустыми, например, если игра была выпущена в конце года.

Столбец user_score (оценка пользователей) - приведем к вещественному типу.Обратим внимание, что столбец user_score имеет строковый тип, а не вещественный. Отсортируем таблицу по столбцу "user_score"
и выясним почему он имеет не правильный тип.

```python
Games.sort_values(by='user_score', ascending=False).reset_index()
```



После сортировки таблицы, видно что некоторые значения столбца user_score имеют строковые значения tbd.
'To Be Determined' - будет определено позже, можно заменить на значение не определено. Чтобы привести столбец к вещественному типу и иметь возможность применить к нему математические операции.   
•	заменим tbd на NaN  
•	приведем столбец user_score к вещественному типу.

```python
Games['user_score'] = Games['user_score'].replace('tbd', np.nan, regex=True)
Games['user_score'] = Games['user_score'].astype('float64')
```

Удалим пустые значения в столбцах name, genre.

```python
Games = Games.dropna(subset=['name', 'genre']).reset_index(drop=True)
Games.info() #Проверка
```

Приведем в порядок данные в столбце year_of_release.

```python
# Посчитаем количество пропущенных значений в столбце 'year_of_release'
Games['year_of_release'].isnull().sum() 
```

```python
# так как название игр может совпадать, то в цикле сделаем замену, вместо NaN, поставим год выпуска
for i in Games[Games['year_of_release'].isnull() == True].index:  
    Games['year_of_release'][i] = Games.loc[Games['name'] == Games['name'][i], 'year_of_release'].max()
```

Выведем количество пустых значении в стобце year_of_release , после применения цикла.

```python
Games['year_of_release'].isnull().sum()
```

Удалим строки с пропущенными значениями из year_of_release

Избавляемся от строк с пропусками

```python
Games = Games.dropna(subset=['year_of_release']).reset_index(drop=True)
```

Так как оценка пользователей и критиков может иметь значение больше или равно нулю. Заменим пропущенные значения в столбцах critic_score и user_score на -1.

```python
Games['critic_score'] = Games['critic_score'].fillna(-1)
Games['user_score'] = Games['user_score'].fillna(-1)
Games.info() #Проверим
```

## Изменения продажи по платформам. Платформы с наибольшими суммарными продажами,  распределение по годам. Определение за какой срок появляются новые и исчезают старые платформы  

```python
# Создадим таблицу платформ, на которой были самые высокие продажи:
platform_on_sales = Games.pivot_table(index='platform', values='total_sales', aggfunc = 'sum')\
                        .sort_values(by='total_sales', ascending=False).reset_index()[:10]
platform_on_sales # Проверка
```

## Оценка влияния внутри одной популярной платформы отзывов пользователей и критиков. Диаграмма рассеяния и корреляция между отзывами и продажами. Соотнесение с продажами игр на других платформах  

Оценим:  
•	Как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков.  
•	Построим диаграмму рассеяния и посчитаем корреляцию между отзывами и продажами.

Проведем анализ для платформы PS4 с максимальной средней выручной за игру в 801 тыс.долл.

```python
#выделим в отдельную таблицу продажи по PS4 за актуальные периоды
game_stat_ps4 = period_df.query('platform == "PS4" and user_score != -1 and critic_score != -1')
```

```python
game_stat_ps4.plot(y='total_sales', x='user_score', kind='scatter', alpha=0.5, title='Зависимость выручки платформы PS4 от оценки пользователей за актуальные периоды')
ax = plt.gca()
ax.set_ylabel('Выручка, млн дол.')
ax.set_xlabel('Оценка пользователей')

game_stat_ps4.plot(y='total_sales', x='critic_score', kind='scatter', alpha=0.5, title='Зависимость выручки платформы PS4 от оценки критиков за актуальные периоды')
ax = plt.gca()
ax.set_ylabel('Выручка, млн дол.')
ax.set_xlabel('Оценка критиков')

print ('Коэффициент корреляции общей выручки платформы PS4 и оценки пользователей', game_stat_ps4['total_sales'].corr(game_stat_ps4['user_score']))
print ('Коэффициент корреляции общей выручки платформы PS4 и оценки критиков', game_stat_ps4['total_sales'].corr(game_stat_ps4['critic_score']))
```

И по графикам, и по значению коэффициента корреляции видно, что отзывы критиков в большей степени влияют на продажи, чем отзывы пользователей. Больше всего выручка при оценке критиков 70 и выше.

### Проверка гипотез


  -           Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
  -           Средние пользовательские рейтинги жанров Action (англ. «действие») и Sports (англ. «виды спорта») разные. Задайте самостоятельно пороговое значение alpha

Достаточно детально проанализировали наши данные и изучили портреты пользователей.
Давай теперь проверим наши гипотезы.


### Средние пользовательские рейтинги платформ Xbox One и PC (по актуальным данным)


   -           H0 - cредние пользовательские рейтинги платформ XboxOne и PC одинаковые  
   -        H1 - cредние пользовательские рейтинги платформ XboxOne и PC разные


```python
hypo=period_df[period_df['platform'].isin(['PC', 'XOne']) ].dropna()
hypo=hypo.query('user_score != -1')
# дисперсии
print( np.var(hypo.query('platform =="XOne"')['user_score']))
print( np.var(hypo.query('platform =="PC"')['user_score']))
```



так как разница дисперсии больше 5%, параметр equal_var = False


```python
alpha = 0.01

results = st.ttest_ind(hypo.query('platform =="XOne"')['user_score'], hypo.query('platform =="PC"')['user_score'], equal_var = False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу, так как cредние пользовательские рейтинги платформ XOne и PC разные.")
else:
    print("Не получилось отвергнуть нулевую гипотезу так как разница между cредними пользовательские рейтинги платформ XOne и PC статистически не значима")
```

Вывод: Нулевая гипотеза не была отвергнута, это значит, что разница между cредними пользовательские рейтинги платформ XOne и PC статистически не значима.

### Средние пользовательские рейтинги жанров Action (англ. «действие») и Sports (англ. «виды спорта»)


   -       H0 - cредние пользовательские рейтинги жанров Action и Sports одинаковые
   -       H1 - cредние пользовательские рейтинги жанров Action и Sports разные


```python
hypo=period_df[period_df['genre'].isin(['Sports', 'Action']) ]
hypo=hypo.query('user_score != -1')
# дисперсии
print( np.var(hypo.loc[hypo['genre']=='Action', 'user_score']))
print( np.var(hypo.loc[hypo['genre']=='Sports', 'user_score']))
```

так как разница дисперсии больше 5%, параметр equal_var = False

```python
alpha = 0.01

results = st.ttest_ind(hypo.query('genre =="Action"')['user_score'],
                       hypo.query('genre =="Sports"')['user_score'], equal_var = False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу, так как cредние пользовательские рейтинги жанров Action и Sports разные.")
else:
    print("Не получилось отвергнуть нулевую гипотезу так как разница между пользовательскими рейтингами жанров Action и Sports статистически не значима")
```

Вывод: Нулевая гипотеза отвергнута - это значит, что и Гипотеза №1 отвергнута, а средние пользовательские рейтинги жанров Action и Sports разные.
