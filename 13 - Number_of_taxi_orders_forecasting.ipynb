```python
 def df_analysis(df, rolling_size=10):

    print(df['num_orders'].describe())

    plt.figure(figsize=(15,15))

    plt.subplot(3,1,1)
    sns.lineplot(data=df, x=df.index, y='num_orders', label='num_orders')
    sns.lineplot(data=df, x=df.index, y=df['num_orders'].rolling(rolling_size).mean(), label='mean')
    sns.lineplot(data=df, x=df.index, y=df['num_orders'].rolling(rolling_size).std(), label='std')
    plt.grid(True)
    plt.legend()

    plt.subplot(3,1,2)
    sns.histplot(df, x='num_orders', kde=True)
    plt.grid(True)

    plt.subplot(3,1,3)
    sns.boxplot(data=df, x='num_orders')
    plt.grid(True)
    plt.show()
```

```python
def df_decomposition(df, day_name=False):    
    decomposed = seasonal_decompose(df)
    plt.figure(figsize=(15,25))
    plt.subplot(3,1,1)
    if day_name:
        sns.lineplot(data=decomposed.trend, x=decomposed.trend.index.strftime('%d %A'), y=decomposed.trend)
    else:
        sns.lineplot(data=decomposed.trend, x=decomposed.trend.index, y=decomposed.trend)
    plt.grid()
    plt.xticks(rotation=25)
    plt.title('Trend/Тренд')
    plt.subplot(3,1,2)
    if day_name:
        sns.lineplot(data=decomposed.seasonal, x=decomposed.trend.index.strftime('%d %A'), y=decomposed.seasonal)
    else:
        sns.lineplot(data=decomposed.seasonal, x=decomposed.trend.index, y=decomposed.seasonal)
    plt.grid()
    plt.xticks(rotation=25)
    plt.title('Seasonality/Сезонность')
    plt.subplot(3,1,3)
    if day_name:
        sns.lineplot(data=decomposed.resid, x=decomposed.trend.index.strftime('%d %A'), y=decomposed.resid)
    else:
        sns.lineplot(data=decomposed.resid, x=decomposed.trend.index, y=decomposed.resid)
    plt.grid()
    plt.xticks(rotation=25)
    plt.title('Residuals/Остаток')
    plt.show()
```

# 3. Обучение моделей

## 3.1. Подготовка данных 

Создадим признаки час, день недели, месяц, отстающие значения и скользящее среднее

```python
def make_features(df, max_lag=10, rolling_window_size=10):
    df_with_features = df.copy()
    df_with_features['dayofweek'] = df_with_features.index.dayofweek
    df_with_features['hour'] = df_with_features.index.hour
    df_with_features['month'] = df_with_features.index.month
    
    for lag in range(1, max_lag + 1):
        df_with_features['lag_{}'.format(lag)] = df_with_features['num_orders'].shift(lag)
    
    df_with_features['rolling_mean'] = df_with_features['num_orders'].shift().rolling(rolling_window_size).mean()
    df_with_features.dropna(inplace=True)

    return df_with_features
```

```python
df_taxi_model = make_features(df_taxi_hour)
```

```python
df_taxi_model.head()
```

```python
X = df_taxi_model.drop('num_orders', axis=1)
y = df_taxi_model['num_orders']

X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=False, test_size=0.1)
```

Проведем проверку соотношения размеров выборок.

```python
X_train.shape[0] / X_test.shape[0]
```

Проведем проверку правильности расположения временных рядов, чтобы начало тестовой выборки следовал за окончанием обучающей

```python
display(X_train.tail(2))
X_test.head(2)
```

```python
tscv = TimeSeriesSplit(n_splits=3)
```

Определим функцию для расчета rmse на кросс-валидации

```python
def rmse_cv(estimator, X=X_train, y=y_train):
    rmse  = (cross_val_score(estimator=estimator, 
                                X=X, 
                                y=y, 
                                cv=tscv,
                                n_jobs=-1, 
                                scoring='neg_mean_squared_error').mean() * -1) ** 0.5

    return rmse
```

```python
def rmse_gscv(estimator, param_grid, X=X_train, y=y_train):
    gsearch = GridSearchCV(estimator=estimator, 
                        param_grid = param_grid, 
                        cv=tscv,
                        n_jobs = -1,
                        scoring='neg_mean_squared_error')
    gsearch.fit(X, y)
    gsearch_best_score = (gsearch.best_score_*-1) ** 0.5
    print ('RMSE:', gsearch_best_score)
    return gsearch
```

## 3.2. Построение и исследование моделей 

### 3.2.1. Модель LinearRegression 

Построим первую модель на значениях max_lag и rolling_window_size, равных 10.

```python
lr = LinearRegression()
rmse_cv(lr)
```

Проверим возможность улучшения модели за счет изменения значений max_lag и rolling_window_size.

```python
list_lag=[]
list_rws=[]
list_rmse=[]
for lag in range(1, 51):
    for rws in range(1, 26):
        df_search = make_features(df_taxi_hour, max_lag=lag, rolling_window_size=rws)
        X_full = df_search.drop('num_orders', axis=1)
        y_full = df_search['num_orders']
        X_temp_train, X_temp_test, y_temp_train, y_temp_test = train_test_split(X_full, y_full, shuffle=False, test_size=0.1)
        estimator=LinearRegression()
        rmse_search = rmse_cv(estimator, X=X_temp_train, y=y_temp_train)
        list_lag.append(lag)
        list_rws.append(rws)
        list_rmse.append(rmse_search)
        
```

```python
plt.figure(figsize=(15,15))
(sns.heatmap(pd.DataFrame({'lag':list_lag, 'rws':list_rws, 'rmse':list_rmse}).
    pivot_table(index='lag', columns='rws', values='rmse'), annot=True, fmt='0.1f'))
```

### Результат
    Из проведенных расчетов можно сделать следующие выводы:
    • качество модели возрастает до достижения max_lag значения 48ч, при этом качество модели скачкообразно возрастает при достижении max_lag показателей в 24 и 48 часов;
    • при небольших значениях max_lag, качество модели несколько возрастает при достижении значения rolling_window_size 24  часа, когда размер max_lag больше 24 часа, качество модели практически не зависит от размера скользящего окна.

Для построения признаков определим max_lag равным 48ч, а размер окна, для скользящего среднего оставим по умолчанию равным 10.

```python
df_taxi_opt = make_features(df_taxi_hour, max_lag=48, rolling_window_size=10)
X_opt = df_taxi_opt.drop('num_orders', axis=1)
y_opt = df_taxi_opt['num_orders']
X_opt_train, X_opt_test, y_opt_train, y_opt_test = train_test_split(X_opt, y_opt, shuffle=False, test_size=0.1)
```

# 4. Тестирование моделей

## 4.1. Проведение тестирования 

Для исследованных моделей произведем обучение на всем наборе обучающих данных, для модели случайного леса и градиентного бустинга с ипользованием подобранных гиперппараметров. Посчиатем длительность обучения, а также качество предсказания на тестовых данных. Подготовим соответствующую функцию.

```python
def model_result(model, X_train, X_test, y_train, y_test):
    
    start1 = time.time()
    model.fit(X_train, y_train)
    end1 = time.time()
    time_train = end1 - start1

    start2 = time.time()
    y_pred = model.predict(X_test)
    end2 = time.time()
    time_predict = end2 - start2

    rmse = mean_squared_error(y_test, y_pred) ** 0.5
    return rmse, time_train, time_predict
```

```python
# LinearRegressor
lr_result = model_result(model=LinearRegression(), 
                                X_train=X_opt_train, 
                                X_test=X_opt_test,
                                y_train=y_opt_train,
                                y_test=y_opt_test)
print('Время обучения', lr_result[1])
print('Скорость предсказания', lr_result[2])   
print('RMSE', lr_result[0]) 
```

```python
# RandomForestRegressor
rfr_result = model_result(model=RandomForestRegressor(random_state=555, **rfr.best_params_), 
                                X_train=X_opt_train, 
                                X_test=X_opt_test,
                                y_train=y_opt_train,
                                y_test=y_opt_test)
print('Время обучения', rfr_result[1])
print('Скорость предсказания', rfr_result[2])   
print('RMSE', rfr_result[0]) 
```

```python
# LightGBMRegressor 
lgbmr_result = model_result(model=LGBMRegressor(random_state=555, **lgbmr.best_params_), 
                                X_train=X_opt_train, 
                                X_test=X_opt_test,
                                y_train=y_opt_train,
                                y_test=y_opt_test)
print('Время обучения', lgbmr_result[1])
print('Скорость предсказания', lgbmr_result[2])   
print('RMSE', lgbmr_result[0]) 
```

```python
model_list = [lr_result, rfr_result, lgbmr_result]
rmse_list = [model[0] for model in model_list]
time_train_list = [model[1] for model in model_list]
time_predict_list = [model[2] for model in model_list]
model_name_list = ['LinearRegresson', 'RandomForestRegressor', 'LightGBMRegressor']
```

```python
results_df = pd.DataFrame(data={'RMSE':rmse_list, 'Time_train':time_train_list, 'Time_predict':time_predict_list,},
                                                                                index=model_name_list)
results_df
```

### Результат
Проведенные расчеты показали, что значение метрики RMSE на тестовой выборке составляет для рассмотренных моделей от 39.455 до 42.586, что соответствует требованиям проекта по качеству моделей. Лучшее значение метрики, а также сравнительно низкое время обучения получено на модели градиентного бустинга при реализации его библиотекой LightGBM. Указанная модель может быть рекомендована для прогноза количества заказов.
